```python
from matplotlib import pyplot as mpl;
import pandas as pd;

data = pd.read_csv("GW150914.dat", sep = "\s+", names = ["Time", "Hanford", "Livingston"]);

#mpl.plot(data["Time"], data["Hanford"], label = "Hanford");
mpl.plot(data["Time"], data["Livingston"], label = "Livingston");

mpl.axvline(-0.04, color = 'red', label = '-0.041');
mpl.axvline(-0.028, color = 'orange', label = '-0.042');
mpl.axvline(-0.032, color = 'yellow', label = '-0.032');
mpl.axvline(-0.014, color = 'green', label = '-0.014');
mpl.axvline(-0.01, color = 'blue', label = '-0.01');

mpl.xlabel("Time (Seconds)");
mpl.ylabel("Gravitational Wave Signal");
mpl.title("Gravitational Wave Observation");
mpl.legend();

mpl.show();
```
